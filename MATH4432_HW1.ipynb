```python
## a
import pandas as pd
import statsmodels.api as sm
from matplotlib import pyplot as plt
from scipy import stats
import seaborn as sns

auto = pd.read_csv('Auto.csv')
auto['horsepower'] = pd.to_numeric(auto['horsepower'], errors ='coerce')
auto.dropna(inplace = True)
auto
```

```python
plt.scatter(auto['horsepower'],auto['mpg'])
```

```python
Y = auto['mpg']
X = auto['horsepower']
X = sm.add_constant(X)
result = sm.OLS(Y, X).fit()
print(result.summary())
```

```python
## (i) - Since the F-statistic is far larger than 1 and the p-value is close to zero,  we reject the null 
##       hypothesis and there is a relationship between horsepower and mpg.
```

```python
## (ii) - The R-squared value of the OLS Regression Results is 0.606, meaning 60.6% of variance in mpg can be
##        explained by horsepower. This is a decently strong relationship.
```

```python
## (iii) - Coefficient = -0.1578, so it is a negative relationship.
```

```python
## (iv) - estimated y = 39.9359 - (0.1578 x 98) = 24.47mpg
##        95% confidence interval - [23.97, 24.96]
##        95% prediction interval - [14.81, 34.12]
```

```python
## b
plt.figure(figsize = (12,8))
sns.regplot(auto['horsepower'], auto['mpg'])
```

```python
## c
plt.figure(figsize = (12,8))
plt.ylim(-20,20)
sns.regplot(result.fittedvalues,result.resid, lowess=True)
plt.axhline(y = 0,linewidth = 0.5,linestyle = 'dashed',color = 'black')
plt.xlabel('Fitted Vales')
plt.ylabel('Residuals')
plt.title('Residual vs Fitted Values')

## U-shape of Residual vs Fitted Values indicates non-linearity in data.
```

```python
ax = stats.probplot(result.resid, dist='norm', plot=plt)
```

```python
## c
predictors = ' + '.join(auto.columns.difference(['name','mpg']))
result = smf.ols('mpg ~ {}'.format(predictors),data = auto).fit()
print(result.summary())
```

```python
## (i)
table = sm.stats.anova_lm(result)
print(table)

## The F-statistic takes a high value (252.4), giving strong evidence to reject the null hypothesis. 
## As a result, there is a relationship between the predictors and response.
```

```python
## (ii) Origin, weight, year and displacement seem to have strong correlations with mpg due to their 
##      low p-values, while acceleration, cylinders and horsepower have p-values above 5%.
```

```python
## (iii) If all other predictors are kept constant, a 1-year time-lapse since the model year will result in an 
## increase of 0.7508 mpg for the car.
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression

ntrain = 50 # Training
ntest = 500 # Test
nrep = 100 # repeat 100 times
p = 20
puse = np.array([1, 2, 3, 4, 10, 20]) # number of predictors
k = np.arange(1 ,10)

sigma = 0.5

Xtrain = np.random.uniform(-1, 1, [ntrain,p])
Xtest = np.random.uniform(-1, 1, [ntest,p])
y0 = np.sin(2 * Xtrain[:, 0]) # Only the first predictor is related to y
ytest = np.sin(2 * Xtest[:, 0])

out_knn = pd.DataFrame() # Output results
out_lm = pd.DataFrame()

for i in np.arange(0, len(puse)):
    yhat_lm = np.zeros((ntest, nrep))
    yhat_knn = np.zeros((ntest, nrep, len(k)))
    
    for l in np.arange(0, nrep):
        y = y0 + np.random.normal(0, sigma, ntrain)
    
        ######### DO: fit linear regression using LinearRegression function, assign predicted value to yhat_lm[:, [l]] #########
        fit_lm = LinearRegression()
        reg = fit_lm.fit(Xtrain[:, 0: puse[i]], y)
        yhat_lm[:, [l]] = np.expand_dims(reg.predict(Xtest[:, 0: puse[i]]), axis = -1)
      
    
        for j in np.arange(1, len(k) + 1):
            ######### DO: fit knn using KNeighborsRegressor function, assign predicted value to yhat_knn[:, l, j] #########
            fit_knn = KNeighborsRegressor(n_neighbors = j)
            knreg = fit_knn.fit(Xtrain[:, 0: puse[i]], y)
            yhat_knn[:, l, j - 1] = knreg.predict(Xtest[:, 0: puse[i]])
        
        
        print(i + 1, "-th p, ", l + 1, "-th repetition finished. \n")
    
  
    ######### DO: compute bias and variance of linear regression #########
    
    # Compute mean of predicted values
    ybar_lm = np.mean(yhat_lm, axis = 1) # E(f^hat)
    
    # Compute bias^2
    biasSQ_lm = np.mean(np.square(ytest - ybar_lm)) # E[ (f - E(f^hat))^2 ]
    
    # Compute variance
    variance_lm = np.mean(np.square(np.expand_dims(ybar_lm, axis = -1) - yhat_lm)) # E[ (E(f^hat) - f^hat)^2 ]
    
    # Compute total MSE
    err_lm = biasSQ_lm + variance_lm
  
    out_lm = out_lm.append(pd.DataFrame({'error': [biasSQ_lm], 'component': ['squared-bias'], 'p': [puse[i]]}))
    out_lm = out_lm.append(pd.DataFrame({'error': [variance_lm], 'component': ['variance'], 'p': [puse[i]]}))
    out_lm = out_lm.append(pd.DataFrame({'error': [err_lm], 'component': ['MSE'], 'p': [puse[i]]}))
    ######### DO: compute bias and variance of knn regression #########
    
    # Compute mean of predicted values
    ybar_knn = np.mean(yhat_knn, axis = 1) # E(f^hat)
    
    # Compute bias^2
    biasSQ_knn = np.mean(np.square(np.expand_dims(ytest, axis = -1) - ybar_knn), axis = 0) # E[ (f - E(f^hat))^2 ]
    
    # Compute variance
    variance_knn = np.mean(np.square(np.expand_dims(ybar_knn, axis = 1) - yhat_knn), axis = (0, 1)) # E[ (E(f^hat) - f^hat)^2 ]
    
    # Compute total MSE
    err_knn = biasSQ_knn + variance_knn
    
    out_knn = out_knn.append(pd.DataFrame({'error': biasSQ_knn, 'component': ['squared-bias'] * len(k), 'K': 1 / np.array(k),
                                       'p': [puse[i]] * len(k)}))
    out_knn = out_knn.append(pd.DataFrame({'error': variance_knn, 'component': ['variance'] * len(k), 'K': 1 / np.array(k),
                                       'p': [puse[i]] * len(k)}))
    out_knn = out_knn.append(pd.DataFrame({'error': err_knn, 'component': ['MSE'] * len(k), 'K': 1 / np.array(k),
                                       'p': [puse[i]] * len(k)}))
```

```python
out_lm
```

```python
fig = plt.figure()
plt.title('Bias-Variance Tradeoff')
plt.xlabel('p')
plt.ylabel('Value')
for frame in [out_lm[out_lm['component'] == 'MSE'], out_lm[out_lm['component'] == 'squared-bias'], 
              out_lm[out_lm['component'] == 'variance']]:
    label = frame['component'][0:1][0]
    plt.plot(frame['p'], frame['error'], '-o', label = label)
    plt.legend()
```
